```python
import cobra
import pandas as pd
import os
import sys
sys.path.append('../')
from modelfunctions import find_group, get_rid

wd = os.path.abspath(os.getcwd()).removesuffix('Code/panGenomeAnalysis')

# read pcm
import_version = 1
pcm = cobra.io.read_sbml_model(f'{wd}Data/pcm/pcm.v{import_version}.xml')
print()

with open(wd + 'Data/analysis/panGenomeAnalysis/Thla_euca_Ara_excl_rxns.txt', 'r') as file:
    rids = [line.strip() for line in file]

gr2nocc = {}
gr2rids = {}
allGenes = set()
for rid in rids:
    for gr in find_group(pcm, rid):
        gr2nocc[gr.name] = gr2nocc.get(gr.name, 0) + 1
        gr2rids[gr] = gr2rids.get(gr, []) + [rid]

    allGenes = allGenes.union(get_rid(pcm, rid).genes)

print(f'{len(allGenes)} genes are responsible for these {len(set(rids))} exclusive reactions')
pd.Series(gr2nocc).sort_values(ascending=False).head()
```
